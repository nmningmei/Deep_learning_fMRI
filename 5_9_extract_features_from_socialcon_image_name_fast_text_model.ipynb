```python
try:
    import fasttext
except:
    !pip install fasttext
    import fasttext
!wget https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.en.300.bin.gz
!gunzip cc.en.300.bin.gz
!rm cc.en.300.bin.gz

model = fasttext.load_model("cc.en.300.bin")

!rm cc.en.300.bin
```

```python
import pandas as pd
df = dict()

for word in words:
    df[word] = model.get_word_vector(word)
df = pd.DataFrame(df)
df.to_csv('fast-text-image-name_features.csv',index = False)
```
